```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
df.isnull().sum()
```

```python
def get_stats():
    temp_df = df.dropna(subset=['Age'])
    df_survived = temp_df[temp_df['Survived'] == 1]
    print('Num survived: ', df_survived.shape[0])

    x_bar = df_survived['Age'].mean()
    print('Mean: ', x_bar)

    s = df_survived['Age'].std()
    print('Standard Deviation: ', s)

    n =  df_survived.shape[0]

    standard_error = s / np.sqrt(n)
    print('Standard Error: ', standard_error)

    return x_bar, s, standard_error, n
```

```python
get_stats()
```

```python
def plothist():
    temp_df = df.dropna(subset=['Age'])
    df_survived = temp_df[temp_df['Survived'] == 1]
    plt.title('Age distribution of survivors')
    sns.histplot(df_survived['Age'], kde=True)
```

```python
plothist()
```

```python
temp_df = df.dropna(subset=['Age'])
df_survived = temp_df[temp_df['Survived'] == 1]

children = df_survived[df_survived['Age'] < 18]

male_children = df_survived.loc[children.groupby('Sex')['Age'].groups['male']]
female_children = df_survived.loc[children.groupby('Sex')['Age'].groups['female']]
```

```python
sns.histplot(male_children['Age'])
plt.xlabel('Age for male children')
```

```python
sns.histplot(female_children['Age'])
plt.xlabel('Age for female children')
```
